### 5. 偏態（Skewness）與峰度（Kurtosis）

```python
import numpy as np
from scipy import stats as st

A= data['oil']

SK=st.skew(A)          #偏態
KU=st.kurtosis(A)      #峰態 
JB=st.jarque_bera(A)   #JB 

print('偏態 =',SK)
print('峰度 =',KU)
print("JB統計值 = {} 機率= {}".format(JB[0], JB[1]))
```

# 平均移動線

# 相關係數import pandas as pd

```python
import numpy as np
from numpy import array, corrcoef

A=data['USA']
B=data['GOLD']
C=data['oil']
D=data['Bitcorn']

# 將資料變成陣列
rho_ALL = array([A, B,C,D])

# 計算兩組數的相關係數
# 返回結果為矩陣，第i行第j列的數據表示第i組數與第j組數的相關係數。對角線為1 
corrcoef(rho_ALL)
```

```python
corr=corrcoef(rho_ALL)
import seaborn as sns
sns.heatmap(corr, square=True ,vmax=1.0, linecolor='white', annot=True)
plt.savefig('array.pdf')
plt.savefig('array.jpg', format='jpg')
```
