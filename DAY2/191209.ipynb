在比特幣問世後的第一次全球金融市場大動盪，即2018年3月開打的美中貿易戰爆發至今，我國市場上投資人紛紛為其手上的熱錢找尋一個較為合適避險的工具所苦。本研究旨在，以當今投資市場環境慣用的傳統避險策略與新興避險工具比較，找尋出較適合當今市場環境變化的避險工具，供投資者進行避險的資金配置參考依據。

# 2.2. 解決的核心關鍵

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas_datareader import data  

# This is a custom matplotlib style that I use for most of my charts
# plt.style.use('tableau10.mplstyle')

data = pd.read_csv('FIGHT.csv') 
data.head()
```

# 2.2.3.1劃出相關係數圖 找到資產組合

```python
import numpy as np
from numpy import array, corrcoef

A=data['USA']
B=data['gold']
C=data['Y9999']
D=data['oil']
E=data['bond']

# 將資料變成陣列
rho_ALL = array([A, B,C,D,E])

# 計算兩組數的相關係數
# 返回結果為矩陣，第i行第j列的數據表示第i組數與第j組數的相關係數。對角線為1 
corrcoef(rho_ALL)
```

```python
corr=corrcoef(rho_ALL)
import seaborn as sns
X=("oil",'USA','gold','Y9999','oil","bond')
sns.heatmap(corr, square=True ,vmax=1.0, linecolor='white',annot=True)
#sns.set(xticklabels=['X'])
plt.savefig('array.pdf')
plt.savefig('array.jpg', format='jpg')
```

# 2.2.3.2 找出相關係數最大的2個標低，並劃出他的股價圖

```python
import numpy as np 
import matplotlib.pyplot as plt 

fig, ax1 = plt.subplots(figsize=(20,5))

plt.title('Graph Drawing of USA & Bitcorn')
XX=B=data['Date']
plt.xlabel('date')
ax2 = ax1.twinx()

B=data['bond']
ax1.set_ylabel('bond', color='blue')
ax1.plot(XX, B, color='blue', alpha=0.75)
ax1.tick_params(axis='y', labelcolor='blue')

U=data['gold']
ax2.set_ylabel('gold', color='red')
ax2.plot(XX, U, color='red', alpha=1)
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()

#plt.savefig('line of USA.pdf')
#plt.savefig('line of USA.jpg', format='jpg')
```

# 2.2.3.3先假設投資比例為0.2和0.8(之後會用最適配置求出)，再畫出報酬

避險，又稱對沖、保值套利，最常見進出口的貿易商，為避免外匯損失，利用遠期外匯來進行操作，是一種能在減低商業風險的同時，能在投資獲利的手法。

```python
import numpy as np 
import matplotlib.pyplot as plt 
X= data['Date']
y1 = data['gold']
y2 = data['bond']
Y=0.2*y1-0.8*y2

fig, ax1 = plt.subplots(figsize=(20,5))

plt.title('Graph Drawing of USA & Bitcorn')
XX=B=data['Date']
plt.xlabel('date')
ax2 = ax1.twinx()


ax1.set_ylabel('GG', color='blue')
ax1.plot(XX, Y, color='blue', alpha=0.75)
ax1.tick_params(axis='y', labelcolor='blue')

U=data['Y9999']
ax2.set_ylabel('Y9999', color='red')
ax2.plot(XX, U, color='red', alpha=1)
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.show()

#plt.savefig('line of USA.pdf')
#plt.savefig('line of USA.jpg', format='jpg')
```
